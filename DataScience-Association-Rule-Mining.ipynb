```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
```

```python
df_csv = pd.read_csv('supermarket.csv', header=None, on_bad_lines='skip')
```

```python
dataset = df_csv.values.tolist()
```

```python
dataset
```

```python
dataset = [list(filter(pd.notna, sublist)) for sublist in dataset]
```

```python
dataset
```

```python
dataset = [list(filter(lambda x: isinstance(x, str) and x.strip() != '', row.dropna().tolist())) for _, row in df_csv.iterrows()]
```

```python
dataset
```

```python
te = TransactionEncoder()
te_ary = te.fit(dataset).transform(dataset)
df = pd.DataFrame(te_ary, columns=te.columns_)
```

```python
import csv
```

```python
data = []
with open('supermarket.csv', 'r', newline='') as csv_file:
    csv_reader = csv.reader(csv_file)
    for row in csv_reader:
        try:
            data.append(row[:])
        except IndexError:
            print(f"Row skipped: {row}")

te = TransactionEncoder()
te_ary = te.fit(data).transform(data)
df = pd.DataFrame(te_ary, columns=te.columns_)
```

```python
df['low']
```

```python
frequent_itemsets = apriori(df, min_support=0.25, use_colnames=True)
```

```python
frequent_itemsets
```

```python
max_length = frequent_itemsets['itemsets'].apply(len).max()
largest_itemsets = frequent_itemsets[frequent_itemsets['itemsets'].apply(len) == max_length]
```

```python
max_length
```

```python
largest_itemsets
```

```python
largest_itemset = largest_itemsets.sort_values(by="support", ascending=False).iloc[0]
```

```python
print(f"Itemset: {largest_itemset['itemsets']}")
print(f"Support: {round(largest_itemset['support'], 2)}")
```

```python
from mlxtend.frequent_patterns import association_rules
```

```python
frequent_itemsets = apriori(df, min_support=0.25, use_colnames=True)

rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.75)

rules[['support', 'confidence', 'lift']] = rules[['support', 'confidence', 'lift']].round(2)

print(rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']])
```
